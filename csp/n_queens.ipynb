```python
N = 16

variables = []
for i in range(N):
    variables.append(i)

domains = {col: list(range(N)) for col in variables}
```

```python
def is_consistent(col, row, assignment):
    for other_col in assignment:
        other_row = assignment[other_col]

        if row == other_row or abs(col - other_col) == abs(row - other_row):
            return False
        
    return True
```

```python
def backtrack(assignment):
    if len(assignment) == len(variables):
        return assignment
    
    col = None
    for v in variables:
        if v not in assignment:
            col = v
            break

    for row in domains:
        if is_consistent(col, row, assignment):
            assignment[col] = row

            result = backtrack(assignment)

            if result is not None:
                return assignment
            
            del assignment[col]

    return None

solution = backtrack({})
solution
```

```python
def print_board(solution, n):
    for row in range(n):
        line = ""
        for col in range(n):
            if solution[col] == row:
                line += " Q"
            else:
                line += " ."
        print(line)

print_board(solution, N)
```
