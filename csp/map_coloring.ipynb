```python
variables = ['A', 'B', 'C', 'D']

domains = {
    "A": ["Red", "Green ", "Blue"],
    "B": ["Red", "Green ", "Blue"],
    "C": ["Red", "Green ", "Blue"],
    "D": ["Red", "Green ", "Blue"]
}

neighbors = [
    ("A", "B"),
    ("A", "C"),
    ("B", "C"),
    ("B", "D")
]

assignment = {"A": "Blue"}
```

```python
def is_consistent(variable, value, assignment):
    for pair in neighbors:
        var1 = pair[0]
        var2 = pair[1]

        if variable == var1:
            if var2 in assignment:
                if assignment[var2] == value:
                    return False

        if variable == var2:
            if var1 in assignment:
                if assignment[var1] == value:
                    return False

    return True
```

```python
def backtrack(assignment):
    if len(assignment) == len(variables):
        return assignment
    
    var = None
    for v in variables:
        if v not in assignment:
            var = v
            break
    
    for value in domains[var]:
        if is_consistent(var, value, assignment):
            assignment[var] = value
            result = backtrack(assignment)

            if result is not None:
                return result
            
            del assignment[var]

    return None

solution = backtrack(assignment)
print(solution)
```
