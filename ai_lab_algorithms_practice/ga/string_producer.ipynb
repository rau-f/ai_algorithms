```python
import random

def create_chromosome(length):
    chrom = []
    
    for i in range(length):
        gene = random.randint(0, 1)
        chrom.append(gene)

    return chrom
```

```python
def fitness(chromosom):
    count = 0

    for gene in chromosom:
        if gene == 1:
            count += 1

    return count
```

```python
def select_parent(population, scores):
    total = 0
    for s in scores:
        total += s

    spin = random.uniform(0, total)

    running_sum = 0
    for i in range(len(population)):
        running_sum += scores[i]
        if spin <= running_sum:
            return population[i]

    return population[-1]
```

```python
def crossover(p1, p2):
    point = random.randint(0, len(p1) - 1)
    
    child1 = p1[:point] + p2[point:]
    child2 = p2[:point] + p1[point:]

    return child1, child2
```

```python
def mutate(chromosom, rate):
    new_chrom = []
    for gene in chromosom:
        if random.random() < rate:
            if gene == 0:
                new_chrom.append(1)
            else:
                new_chrom.append(0)
        else:
            new_chrom.append(gene)

    return new_chrom
```

```python
CHROM_LEN      = 9
POP_SIZE       = 20
MAX_GENS       = 50
MUTATION_RATE  = 0.05
CORSSOVER_RATE = 0.8

population = []

for i in range(POP_SIZE):
    population.append(create_chromosome(CHROM_LEN))

for gen in range(MAX_GENS):
    scores = []

    for chrom in population:
        scores.append(fitness(chrom))

    best_score = 0
    best_chrom = population[0]

    for i in range(POP_SIZE):
        if scores[i] > best_score:
            best_score = scores[i]
            best_chrom = population[i]

    print("Gen", gen, "| Best fitness:", best_score, "| Chromosom:", best_chrom)

    new_population = []

    while len(new_population) < POP_SIZE:
        p1 = select_parent(population, scores)
        p2 = select_parent(population, scores)

        if random.random() < CORSSOVER_RATE:
            child1, child2 = crossover(p1, p2)
        else:
            child1 = p1.copy()
            child2 = p2.copy()

        mutate(child1, MUTATION_RATE)
        mutate(child2, MUTATION_RATE)

        new_population.append(child1)
        new_population.append(child2)

    population = new_population[:POP_SIZE]

print()
print("Final population's best chromosome :", best_chrom)
print ("Its fitness :", best_score )
```
