```python
import heapq

def a_star(graph, src, dst, h):
    minHeap = [(0, 0, src, 'NULL')]
    visited = {}

    while minHeap:
        _, c1, n1, prev = heapq.heappop(minHeap)

        if n1 in visited:
            continue
        visited[n1] = (prev, c1)

        if n1 == dst:
            break

        for n2, c2 in graph[n1]:
            if n2 not in visited:
                heapq.heappush(minHeap, (c1 + c2 + h[n2], c1 + c2, n2, n1))

    path = []
    node = dst

    while node != src:
        path.append(node)
        node = visited[node][0]

    path.append(src)
    path.reverse()

    return path, visited[dst][1]

a_star(romania_map, 'Arad', 'Bucharest', sld_to_bucharest)
```
