```python
n = 8       # nodes count
E = [       # edges
    [0, 1],
    [0, 3],
    [1, 2],
    [3, 4],
    [3, 6],
    [3, 7],
    [4, 2],
    [4, 5],
    [5, 2],
]
```

# Edges -> Adjecency Matrix

```python
# adjecency matrix
M = []

for i in range(n):
    M.append([0] * n)

for u, v in E:
    M[u][v] = 1

    # for undirected graph [diagonal matrix]
    # M[v][u] = 1

M
```

# Edges to Adjecency List

```python
from collections import defaultdict

D = defaultdict(list)

for u, v in E:
    D[u].append(v)

    # for undirected graph
    # D[v].append(u)

D
```

# DFS

```python
def dfs_recursive(node):
    print(node)
    for nei_node in D[node]:
        if nei_node not in seen:
            seen.add(nei_node)
            dfs_recursive(nei_node)

source = 0
seen = set()
seen.add(source)

dfs_recursive(source)
```

```python
def dfs_iterative(D):
    source = 0
    seen = set()
    seen.add(source)
    stack = [source]

    while stack:
        node = stack.pop()
        print(node)

        for nei_node in D[node]:
            if nei_node not in seen:
                seen.add(nei_node)
                stack.append(nei_node)

dfs_iterative(D)
```

# BFS

```python
from collections import deque

def bfs(D):
    source = 0
    seen = set()
    queue = deque()

    seen.add(source)
    queue.append(source)

    while queue:
        node = queue.popleft()
        print(node)

        for nei_node in D[node]:
            if nei_node not in seen:
                seen.add(nei_node)
                queue.append(nei_node)

bfs(D)
```
