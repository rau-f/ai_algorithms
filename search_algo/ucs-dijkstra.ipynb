```python
import heapq

def dijkstra(graph, src, dst):
    minHeap = [(0, src, 'NULL')]
    shortest = {}
    
    while minHeap:
        c1, n1, prev = heapq.heappop(minHeap)

        if n1 in shortest:
            continue

        shortest[n1] = (c1, prev)

        if n1 == dst:
            break

        for n2, c2 in graph[n1]:
            if n2 not in shortest:
                heapq.heappush(minHeap, (c1 + c2, n2, n1))

    path = []
    node = dst

    while node != src:
        _, prev = shortest[node]
        path.append(node)
        node = prev
        
    path.append((src, 0))
    path.reverse()

    return path, shortest[dst][0]


dijkstra(romania_map, 'Arad', 'Bucharest')
```
